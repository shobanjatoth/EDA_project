# 1) CLEAN NUMERIC COLUMNS

```python


df["Selling Price"] = (
    df["Selling Price"].astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)
df["Selling Price"] = pd.to_numeric(df["Selling Price"], errors="coerce")

df["MRP"] = (
    df["MRP"].astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)
df["MRP"] = pd.to_numeric(df["MRP"], errors="coerce")

df["Discount"] = (
    df["Discount"].astype(str)
    .str.replace("% OFF", "", regex=False)
    .str.replace("%", "", regex=False)
    .str.strip()
)
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")

df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")

# Derived business columns
df["Discount_Amount"] = df["MRP"] - df["Selling Price"]
df["Discount_Computed_%"] = round(
    (df["Discount_Amount"] / df["MRP"]) * 100, 1
)

```

```python


def extract_features(title):
    title = str(title)
    title_lower = title.lower()

    # Gender (FIXED)

    if "women" in title_lower or "women's" in title_lower or "womens" in title_lower:
        gender = "Women"
    elif "men" in title_lower or "men's" in title_lower or "mens" in title_lower:
        gender = "Men"
    else:
        gender = "Unisex"

    # Color
    colors = [
        "black","white","blue","green","beige","brown","grey","gray","pink",
        "red","navy","olive","purple","maroon","orange","yellow","teal",
        "lavender","charcoal","cream","off white"
    ]
    color = next((c.title() for c in colors if c in title_lower), None)

    # Fit
  
    fits = ["oversized", "slim fit", "regular", "baggy", "super loose", "boyfriend"]
    fit = next((f.title() for f in fits if f in title_lower), None)

    # Product Type
   
    product_types = [
        "t-shirt","hoodie","sweatshirt","shirt","jacket","jeans","joggers",
        "pants","shorts","vest","pyjamas","boxers","sweater","co-ordinates",
        "track pants","windcheater"
    ]
    product_type = next((p.title() for p in product_types if p in title_lower), None)

    # Print Type
    
    if "graphic" in title_lower:
        print_type = "Graphic"
    elif "typography" in title_lower:
        print_type = "Typography"
    elif "printed" in title_lower:
        print_type = "Printed"
    else:
        print_type = "Solid"

    
    # Wash Type

    if "acid wash" in title_lower:
        wash_type = "Acid Wash"
    elif "tie & dye" in title_lower or "tie and dye" in title_lower:
        wash_type = "Tie & Dye"
    else:
        wash_type = None

  
    # Theme / Collab
   
    themes = ["marvel","batman","garfield","anime","naruto","one piece","mickey"]
    theme = next((t.title() for t in themes if t in title_lower), None)

    return pd.Series([
        gender, color, fit, product_type, print_type, wash_type, theme
    ])


# APPLY FEATURE EXTRACTION

df[[
    "Gender", "Color", "Fit", "Product_Type",
    "Print_Type", "Wash_Type", "Theme"
]] = df["Title"].apply(extract_features)



# CATEGORY MAPPING

category_map = {
    "T-Shirt": "Topwear",
    "Hoodie": "Outerwear",
    "Sweatshirt": "Outerwear",
    "Shirt": "Topwear",
    "Jacket": "Outerwear",
    "Jeans": "Bottomwear",
    "Joggers": "Bottomwear",
    "Pants": "Bottomwear",
    "Shorts": "Bottomwear",
    "Vest": "Topwear",
    "Pyjamas": "Loungewear",
    "Boxers": "Loungewear",
    "Sweater": "Outerwear",
    "Co-Ordinates": "Set",
    "Track Pants": "Bottomwear",
    "Windcheater": "Outerwear"
}

df["Category"] = df["Product_Type"].map(category_map)
```
