```python
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

base_url = "https://www.bewakoof.com/men-clothing?page="

product_names = []
selling_price = []
mrp = []
discount = []
ratings = []
offers = []

# Bewakoof has ~150 pages (20 products per page)
for page in range(1, 160):   # 1 → 159
    print("Scraping page:", page)

    driver.get(base_url + str(page))
    time.sleep(4)

    # Wait until products load
    if not driver.find_elements(By.XPATH, "//span[@data-testid='product-title']"):
        print("No more products")
        break

    # ---- YOUR ORIGINAL CODE (unchanged) ----

    for tag in driver.find_elements(By.XPATH, "//span[@data-testid='product-title']"):
        text = tag.text
        a = re.findall(r'(.+)', text)
        product_names.extend(a)

    for tag in driver.find_elements(By.XPATH, "//span[@data-testid='product-card-selling-price']"):
        text = tag.text
        a = re.findall(r'₹(.+)', text)
        selling_price.extend(a)

    for tag in driver.find_elements(By.XPATH, "//span[@data-testid='product-card-mrp']"):
        text = tag.text
        a = re.findall(r'₹(.+)', text)
        mrp.extend(a)

    for tag in driver.find_elements(By.XPATH, "//span[@data-testid='product-card-discount']"):
        text = tag.text
        a = re.findall(r'(\d+)%', text)
        discount.extend(a)

    for tag in driver.find_elements(By.XPATH, "//span"):
        text = tag.text
        a = re.findall(r'^[1-5]\.\d$', text)
        ratings.extend(a)

    for tag in driver.find_elements(By.XPATH, "//span[contains(text(),'BUY')]"):
        text = tag.text
        a = re.findall(r'BUY\s*\d+\s*FOR\s*\d+', text)
        offers.extend(a)

    print("Total so far:", len(product_names))

driver.quit()

# SAVE
df = pd.DataFrame({
    "Product": product_names,
    "Price": selling_price,
    "MRP": mrp,
    "Discount": discount,
    "Rating": ratings[:len(product_names)],
    "Offer": offers
})

df.to_csv("bewakoof_3000_products.csv", index=False)
print("Saved bewakoof_3000_products.csv")

```

```python
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# SPEED UP CHROME
chrome_options = Options()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--blink-settings=imagesEnabled=false")

prefs = {
    "profile.managed_default_content_settings.images": 2,
    "profile.default_content_setting_values.notifications": 2
}
chrome_options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(options=chrome_options)

base_url = "https://www.bewakoof.com/men-clothing?page="
rows = []

for page in range(1, 200):
    print("Scraping page:", page)

    driver.get(base_url + str(page))

    # Wait ONLY until product titles load
    t0 = time.time()
    while time.time() - t0 < 6:
        if driver.find_elements(By.XPATH, "//span[@data-testid='product-title']"):
            break
        time.sleep(0.2)

    products = driver.find_elements(By.XPATH, "//span[@data-testid='product-title']/ancestor::div[2]")

    if not products:
        print("No more products")
        break

    for product in products:
        try:
            title = product.find_element(By.XPATH, ".//span[@data-testid='product-title']").text
        except:
            title = ""

        try:
            sp = re.findall(r'₹(\d+)', product.find_element(By.XPATH, ".//span[@data-testid='product-card-selling-price']").text)[0]
        except:
            sp = ""

        try:
            mrp1 = re.findall(r'₹(\d+)', product.find_element(By.XPATH, ".//span[@data-testid='product-card-mrp']").text)[0]
        except:
            mrp1 = ""

        try:
            disc = re.findall(r'(\d+)%', product.find_element(By.XPATH, ".//span[@data-testid='product-card-discount']").text)[0]
        except:
            disc = ""

        try:
            rating1 = re.findall(r'[1-5]\.\d', product.text)[0]
        except:
            rating1 = ""

        try:
            offer1 = re.findall(r'BUY\s*\d+\s*FOR\s*\d+', product.text)[0]
        except:
            offer1 = ""

        rows.append([title, sp, mrp1, disc, rating1, offer1])
    print("Total collected:", len(rows))

driver.quit()

df = pd.DataFrame(rows, columns=["Title","Selling Price","MRP","Discount %","Rating","Offer"])
df.to_csv("bewakoof_3000_products.csv", index=False)
print("Saved bewakoof_3000_products.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# -------------------- SETUP --------------------
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

url = "https://www.bewakoof.com/winter-wear-for-men"
driver.get(url)

wait = WebDriverWait(driver, 10)

# wait until first product loads
wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, '[data-testid="product-card-info-wrapper"]')
    )
)

# -------------------- AUTO SCROLL --------------------
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2.5)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# -------------------- DATA EXTRACTION --------------------
product_cards = driver.find_elements(
    By.XPATH,
    "//div[@data-testid='product-card-info-wrapper']/ancestor::div[contains(@class,'ProductCard')]"
)

print("Total products found:", len(product_cards))

data = []

for card in product_cards:

    def safe_text(xpath):
        try:
            return card.find_element(By.XPATH, xpath).text
        except:
            return None

    rating = safe_text(".//span[contains(text(),'⭐')]")
    if rating:
        rating = rating.replace("⭐", "").strip()

    data.append({
        "title": safe_text(".//span[@data-testid='product-title']"),
        "price": safe_text(".//span[@data-testid='product-card-selling-price']"),
        "mrp": safe_text(".//span[@data-testid='product-card-mrp']"),
        "discount": safe_text(".//span[@data-testid='product-card-discount']"),
        "rating": rating
    })

driver.quit()

# -------------------- SAVE TO CSV --------------------
df = pd.DataFrame(data).drop_duplicates()
df.to_csv("bewakoof_winter_wear_men.csv", index=False, encoding="utf-8-sig")

print(f"Saved {len(df)} products to bewakoof_winter_wear_men.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# ---------------- SETUP ----------------
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

url = "https://www.bewakoof.com/winter-wear-for-men"
driver.get(url)

wait = WebDriverWait(driver, 15)

wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, '[data-testid="product-card-info-wrapper"]')
    )
)

# ---------------- AUTO SCROLL ----------------
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# ---------------- DATA EXTRACTION ----------------
products = driver.find_elements(
    By.CSS_SELECTOR,
    '[data-testid="product-card-info-wrapper"]'
)

print("Total products found:", len(products))

data = []

for product in products:

    def safe_text(by, value):
        try:
            return product.find_element(by, value).text
        except:
            return None

    title = safe_text(By.CSS_SELECTOR, '[data-testid="product-title"]')
    price = safe_text(By.CSS_SELECTOR, '[data-testid="product-card-selling-price"]')
    mrp = safe_text(By.CSS_SELECTOR, '[data-testid="product-card-mrp"]')
    discount = safe_text(By.CSS_SELECTOR, '[data-testid="product-card-discount"]')

    # ⭐ rating is OUTSIDE info-wrapper → use JS to search parent
    try:
        rating = product.find_element(
            By.XPATH,
            "../..//span[contains(text(),'⭐')]"
        ).text
        rating = rating.replace("⭐", "").strip()
    except:
        rating = None

    data.append({
        "title": title,
        "price": price,
        "mrp": mrp,
        "discount": discount,
        "rating": rating
    })

driver.quit()

# ---------------- SAVE CSV ----------------
df = pd.DataFrame(data).drop_duplicates()
df.to_csv("bewakoof_winter_wear_menjatoth.csv", index=False, encoding="utf-8-sig")

print(f"Saved {len(df)} products to bewakoof_winter_wear_men.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.bewakoof.com/winter-wear-for-men")
time.sleep(5)

product_links = set()
last_height = 0

while True:
    cards = driver.find_elements(By.XPATH, "//a[contains(@href,'/men')]")
    for c in cards:
        link = c.get_attribute("href")
        if link:
            product_links.add(link)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

driver.quit()

print("Total product links:", len(product_links))
```
